```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import date
```

```python
df = pd.read_csv('data_music_trends.csv', sep=",", encoding='cp1252')
```

```python
df.head()
```

```python
sns.pairplot(data=df,
        vars=["bpm", "nrgy", 'dnce', 'dur'],
        kind='scatter',
        hue='year',
        palette='RdBu',
        diag_kind= 'kde'
)

plt.show()
```

```python
# Build a pairplot with different x and y variables
sns.pairplot(data=df,
        x_vars=['bpm', 'val'],
        y_vars=['bpm'],
        kind='scatter',
        hue='year',
        palette='husl',
        diag_kind= 'kde')

plt.show()
```

```python
sns.pairplot(data=df,
        vars=["bpm", "nrgy"],
        kind='scatter',
        palette='RdBu',
        diag_kind= 'kde'
)

plt.show()
```

```python
# Build a JointGrid comparing humidity and total_rentals
sns.set_style("whitegrid")
g = sns.JointGrid(x="bpm",
            y="nrgy",
            data=df,
            xlim=(0, 250)) 

g.plot(sns.regplot, sns.distplot)

plt.show()
```

```python
# Create a jointplot similar to the JointGrid 
sns.jointplot(x="bpm",
        y="nrgy",
        kind='reg',
        order= 2,
        data=df)

plt.show()
plt.clf()
```

```python
# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = (sns.jointplot(x="bpm",
             y="nrgy",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
```

```python
ex1= df1.groupby('year')[['bpm', 'val', 'acous']].max()
```

```python
ex1
```

```python
fig, ax = plt.subplots()

# Plot a bar-chart of gold medals as a function of country
ax.bar(ex1.index, ex1['val'])

# Set the x-axis tick labels to the country names
ax.set_xticklabels(ex1.index, rotation= 90)

# Set the y-axis label
ax.set_ylabel('Valence')

plt.show()
```

```python
# Add bars for "Gold" with the label "Gold"
ax.bar(ex1.index, ex1['bpm'], label= 'tempo')

# Stack bars for "Silver" on top with label "Silver"
ax.bar(ex1.index, ex1['val'], bottom=ex1['bpm'], label= 'Valence')

# Stack bars for "Bronze" on top of that with label "Bronze"
ax.bar(ex1.index, ex1['acous'], bottom=ex1['bpm'] + ex1['val'], label= 'Acousticness')

# Display the legend
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

# Add Seattle temperature data in each month with error bars
ax.errorbar(ex1.index, ex1['bpm'], yerr= ex1['val'].min())

# Add Austin temperature data in each month with error bars
ax.errorbar(ex1.index, ex1['acous'], yerr= ex1['val'].min()) 

# Set the y-axis label
ax.set_ylabel('Temperature (Fahrenheit)')

plt.show()
```
